```python
#Activity 1: Create empty dataframe and add records
#Import the library
import pandas as pd
#Create an empty data frame with column names
df = pd.DataFrame(columns = ['company','model','year'])
#Add records
df.loc[0] = ['Tata', 'Nexon', 2017]
df.loc[1] = ['MG', 'Astor', 2021]
df.loc[2] = ['KIA', 'Seltos', 2019]
df.loc[3] = ['Hyundai', 'Creta', 2015]
#Print the dataframe
df
```

```python
#Activity 2: Create dataframe using numpy array
#Import the library
import numpy as np
# Pass a 2D numpy array - each row is the corresponding row in the dataframe
data = np.array([['Tata','Nexon',2017],
['MG','Astor',2021],
['KIA','Seltos', 2019],
['Hyundai','Creta',2015]])
# pass column names in the columns parameter of the constructor
df = pd.DataFrame(data, columns = ['company', 'model','year'])
df
```

```python
#Activity 3: Create dataframe using list
#Each element of the list is one record in the dataframe. The column headers are passed separately.
# Import pandas library
import pandas as pd
# Create a list of lists
data = [['Tata','Nexon',2017],['MG','Astor',2021],['KIA','Seltos', 2019],['Hyundai','Creta',2015]]
# Create the pandas DataFrame
df = pd.DataFrame(data, columns = ['company', 'model','year'])
# print the dataframe.
df
```

```python
#Activity 4: Create dataframe using dictionary with lists
#Import the library
import pandas as pd
#Create the dictionary
#Lists: Each value associated with the keys in the dictionary data is a list.

data = {'company': ['Tata','MG','KIA','Hyundai'],
'model':['Nexon','Astor','Seltos','Creta'],
'year': [2017,2021,2019,2015]}
#Create the dataframe
df = pd.DataFrame.from_dict(data)
#Print the dataframe
df
```

```python
#Activity 5: Create dataframe using list of dictionary
import pandas as pd
#Each dictionary is a record in the dataframe.
#Dictionary Keys become Column names in the dataframe. Dictionary values become the values of columns
data = [{'company':'Tata','model':'Nexon','year':2017},
{'company':'MG','model':'Astor','year':2021},
{'company':'KIA','model':'Seltos','year':2019},
{'company':'Hyundai','model':'Creta','year':2015}]
df = pd.DataFrame(data)
df
```

```python
#Activity 6: Create dataframe using zip() function
import pandas as pd
#Company list
company = ['Tata','MG','KIA','Hyundai']
#Model list
model = ['Nexon','Astor','Seltos','Creta']
#Year list
year = [2017,2021,2019,2015]
# and merge them by using zip().
data = list(zip(company, model,year))
# Create pandas Dataframe using data
df = pd.DataFrame(data,columns = ['company','model','year'])
# Print dataframe
df
```

```python
#Assignmet 1Set A7
import pandas as pd 
import matplotlib.pyplot as plt
Name = ['rohit', 'raj', 'om', 'kartik', 'jayesh', 'vinit', 'yash', 'pradip', 'tanmay', 'umed']
Percentage = [67.5, 66.6, 77, 78.7, 78, 89, 87.7, 88, 65, 66]
df = pd.DataFrame({'Name': Name, 'Percentage': Percentage})
plt.figure(figsize=(12,6))
ax = df.plot(x="Name", y="Percentage", kind='line')
plt.xlabel("Name")
plt.ylabel("Percentage")
plt.ylim(1, 100)
ax.set_xticks(range(len(Name)))
ax.set_xticklabels(Name, rotation=30, ha="right")
plt.tight_layout()
plt.show()
```

```python
# B-4 
#Write a Python program to get the number of observations, missing values and nan values.

import pandas as pd
import numpy as np


data = {
    'Name': ['Alice', 'Bob', 'Charlie', None],
    'Age': [25, None, 35, 40],
    'Height': [165.4, 170.2, None, 174.0],
    'Graduated': [True, False, True,None ]
}
df = pd.DataFrame(data)
# Printing results
print(f"Number of observations:",df.shape[0])
print(f"Number of missing values:",df.isnull().sum().sum())
print(f"Number of NaN values:",df.isna().sum().sum())
```

```python
# B-5
import pandas as pd
data = {
    'Name': ['Rohit', 'Bob', 'Charlie', 'David'],
    'Age': [21, 30, 35, 40],
    'Height': [1.78, 1.70, 1.68, 1.74],  # Height in meters
    'Weight': [68, 70, 68, 80]  # Weight in kilograms
}
df = pd.DataFrame(data)
# Calculate BMI Body Mass Index
df['BMI'] = df['Weight'] / (df['Height'] ** 2)
print(df)
```

```python
# B-6
import pandas as pd
data = {
    'Name': ['rohit', 'Bob', 'Charlie', 'David'],
    'Age': [21, 30, 35, 40],
    'Height': [1.78, 1.70, 1.68, 1.74],  # Height in meters
    'Weight': [68, 70, 68, 80]  # Weight in kilograms
}
df = pd.DataFrame(data)
# Calculate BMI Body Mass Index
df['BMI'] = df['Weight'] / (df['Height'] ** 2)

# Print the results
print(f"Maximum BMI:",df['BMI'].max())
print(f"Minimum BMI:", df['BMI'].min())
```
